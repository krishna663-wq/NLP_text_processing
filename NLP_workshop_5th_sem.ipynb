```python
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk


nltk.download('punkt_tab')

corpus=[
    "i love machine learning.",
    "machiine learning is fascinating.",
    "word2vec is an amazing technique for nlp.",
    "i enjoy learning about deep learning."
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, sg=1)

print("vector of 'machine':")
print(model.wv['machine'])

similar_words = model.wv.most_similar('machine', topn=3)
print("\nMost similar words to 'machine':")
for word,similarity in similar_words:
  print(f"{word}:{similarity}")
```

```python
import re
def N_Gram(text,n):
  tokens=re.split('\\s+',text)
  ngram=[]
  for i in range(len(tokens)-n+1):
    temp=[tokens[j] for j in range(i,i+n)]
    ngram.append(" ".join(temp))
  return ngram
```

```python
import re

def remove_mentions_hastags(text):
  text=re.sub(r'[@]','',text) #remove mentions
  text=re.sub(r'#\w+','',text) #remove hashtags
  return text

df['Text'] = df['Text'].apply(remove_mentions_hastags)
```

```python
df['Text'].value_counts()
```

```python
def remove_urls(text):
  text = re.sub(r'https?://\S+|www\.\S+', '', text)
  return text
df['Text'] = df['Text'].apply(remove_urls)
```

```python
def remove_puncuation(text):
  text = re.sub(r'[^\w\s]', '', text)
  return text
df['Text'] = df['Text'].apply(remove_puncuation)
```

```python
def remove_numbers(text):
  text = re.sub(r'\d+', '', text)
  return text
df['Text'] = df['Text'].apply(remove_numbers)
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```
